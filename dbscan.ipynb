```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

moons=make_moons(n_samples=100, shuffle=True,noise=0.05)

plt.scatter(moons[0][:,0],moons[0][:,1])
```

```python
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns

moons=make_moons(n_samples=100, shuffle=True,noise=0.05)
moons=pd.DataFrame(moons[0],columns=(['X1','X2']))

km_fit=KMeans(n_clusters=2)
km_fit.fit(moons[['X1','X2']])
```

```python
moons['cluster']=km_fit.predict(moons[['X1','X2']])
sns.scatterplot(data=moons,x='X1',y='X2',hue='cluster',palette='Spectral')
```

```python
from sklearn.cluster import DBSCAN

moons=make_moons(n_samples=1000, shuffle=True,noise=0.2)
dbscan=DBSCAN(eps=0.1, min_samples=5).fit(moons[0])
#dbscan.labels_
```

```python
sns.scatterplot(data=moons[0],x=moons[0][:,0],y=moons[0][:,1],hue=dbscan.labels_,palette='Spectral')
```
