```python
#shift+Enter
import pandas as pd
import requests #para manejar el protocolo http
from bs4 import BeautifulSoup #es para traerse el html de páginas
```

```python
wiki="https://en.wikipedia.org/wiki/List_of_cities_in_India_by_population"
respuesta=requests.get(wiki)
```

```python
type(respuesta.text)
```

```python
soup=BeautifulSoup(respuesta.text,'html.parser')
```

```python
type(soup)
```

```python
tabla=soup.find('table',{'class':"wikitable"})
```

```python
tabla
```

```python
df=pd.read_html(str(tabla))
df=pd.DataFrame(df[0])
```

```python
df
```
