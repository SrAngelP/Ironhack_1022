```python
mapa_2=folium.Map(location=[19.366399, -99.167778],
                  zoom_start=25)
#Usando el plugin para tener un minimapa

minimapa=plugins.MiniMap(toggle_display=True)

mapa_2.add_child(minimapa)
```

```python
dir1=geocoder.osm('Avenida Coyoacán 703, 03100, Colonia del Valle Sur')
dir2=geocoder.osm('53 Calle Amores, 03100, Colonia del Valle Sur')

mapa_3=folium.Map(location=[19.366399, -99.167778],
                  zoom_start=15)
folium.Marker([dir1.lat,dir1.lng],popup='Fiesta en casa de Maik').add_to(mapa_3)
folium.Marker([dir2.lat,dir2.lng],popup='Mamas pizza').add_to(mapa_3)
mapa_3
```

```python
df_airports=vds.airports()
df_airports.head()
```

```python
mapa=folium.Map(location=[31,-89],zoom_start=6)

for i in range (100):
  folium.Marker(location=[df_airports['latitude'][i],df_airports['longitude'][i]],
                popup=df_airports['iata'][i],tooltip=df_airports['name'][i]).add_to(mapa)
mapa
```

```python
cdmx=geocoder.osm('Mexico city, Mexico')
mty=geocoder.osm('Monterrey, Nuevo Leon')
yuc=geocoder.osm('Merida, Yucatan')

ciudad_calor=[[cdmx.lat,cdmx.lng, 5000],[mty.lat,mty.lng,3000],[yuc.lat,yuc.lng,100] ]

mapa_calor=folium.Map([19,-99],zoom_start=5)
plugins.HeatMap(ciudad_calor,radius=30).add_to(mapa_calor)

mapa_calor
```
