```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data_frame_liquors_2 (2021_08_13 15_24_06 UTC).csv')
```

```python
df.head()
```

```python
df['date'].unique()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df['date'].head()
```

```python
df.info()
```

```python
df['semana']=df['date'].dt.week
df.head()
```

```python
df['date'].dt.year.unique()
```

```python
df_week=df.groupby('semana').agg({'sale_usd':'sum','bottles_sold':'sum'})
df_week
```

```python
df_week=df_week.iloc[2:,].reset_index()
df_week
```

```python
plt.plot(df_week['semana'],df_week['bottles_sold']) #plt.plot(x,y)
```

```python
#realizando la gráfica correspondiente
x=df_week['semana']
y=df_week['bottles_sold']

plt.plot(x,y,'^',c='black')

```

```python
z=df_week['sale_usd']
plt.figure(figsize=(10,5))
plt.title('Número de botellas vendidas por semana',fontsize='25')
plt.ylabel('Botellas vendidas',fontsize='18')
plt.xlabel('Semana',fontsize='15')
plt.scatter(x,y,s=z/10000,c=z,cmap='seismic')
plt.colorbar()
```

```python
top10=df.groupby('item_description',as_index=False).agg({'sale_usd':'sum','bottles_sold':'sum'})
```

```python
top10
```

```python
top10=top10.nlargest(10,'sale_usd')
top10
```

```python
x=top10['item_description']
y=top10['sale_usd']

plt.style.use('ggplot')
plt.figure(figsize=(10,5))
plt.title('Top 10 Botellas',fontsize='18')
plt.xlabel('Botella',fontsize='18')
plt.ylabel('Ventas[USD]',fontsize='14')
plt.plot(x,y,marker='x',linestyle='dashed',markersize=12,color='red')
plt.xticks(rotation='vertical')
#plt.grid()
#1e5= x10⁵ 1.4e5=140000.
```

```python
plt.style.use('default')
plt.figure(figsize=(10,5))
plt.title('Top 10 Botellas',fontsize='18')
plt.xlabel('Botella',fontsize='18')
plt.ylabel('Ventas[USD]',fontsize='14')
plt.plot(x,y,marker='x',linestyle='dashed',markersize=12,color='red')
plt.xticks(rotation='vertical')
```

```python
x=top10['item_description']
y1=top10['sale_usd']
y2=top10['bottles_sold']

plt.style.use('ggplot')

fig,ax=plt.subplots(2)


ax[0].title.set_text('Ventas [USD]')
ax[0].plot(x,y1)
ax[0].set_xticklabels([])


ax[1].title.set_text('Botellas vendidas')
ax[1].scatter(x,y2,c='green')
plt.xticks(rotation=-45,ha='left')
```

```python
plt.plot(x,y1)
plt.show()
print('   ')
plt.plot(x,y2)
plt.show()
```

```python
plt.plot(x,y1)
plt.scatter(x,y2)
```

```python
## gráficas tipo pastel
x=top10['item_description']
y1=top10['sale_usd']
colors=['#1db954','black']  
exp=[0.5,0,0.,0,0.,0.,0,0.25,0,0]
plt.style.use('ggplot')
_, _, autotexts=plt.pie(y1,labels=x,startangle=90,shadow=True,explode=exp,colors=colors,autopct='%1.2f%%')
plt.setp(autotexts,**{'color':'white'})

```

```python
## barplots

plt.figure(figsize=(10,5))
plt.title('Top 10 Botellas',fontsize='18')
plt.xlabel('Botella',fontsize='18')
plt.ylabel('Ventas[USD]',fontsize='14')
plt.xticks(rotation=45, ha='right')

ejex_num=np.arange(len(x))
plt.bar(ejex_num+0.2,y1,0.4,label='Tienda1')
plt.bar(ejex_num-0.2,y2,0.4,label='Tienda2')
plt.xticks(ejex_num,x)
plt.legend(loc='center') #si quieres ver las legendas de cada gráfica, sí o sí hay que poner plt.legend()
```
